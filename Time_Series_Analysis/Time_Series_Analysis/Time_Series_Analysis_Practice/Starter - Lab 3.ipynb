```python
# Load Pandas
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

# Lab 1

```python
# Load the data
df = pd.read_csv("department-sales.csv", index_col= 0, parse_dates=True)
df.head()
```

```python
# Change date format to yyyy-mm-dd
df.index = pd.to_datetime(df.index, format="%d/%m/%Y")
df.head()
```

```python
# Set index frequency to weekly
df.index.freq = "W-FRI"
```

# Lab 2

```python
# Build a simple Data Visualization with matplotlib
plt.figure(figsize=(12,8))
plt.plot(df.Weekly_Sales)
plt.title("Weekly Sales")
plt.show()
```

```python
# Include a vertical line when IsHoliday is True
plt.figure(figsize=(12,8))
plt.plot(df.Weekly_Sales)

for date, row in df.iterrows():
    if row['IsHoliday']:
        plt.axvline(date, color='red', linestyle='--', lw=1)

plt.title("Weekly Sales")
plt.show()
```
